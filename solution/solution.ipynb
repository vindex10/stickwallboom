Сила приложенная к твердому телу разгоняет его центр масс. А момент внешних сил приложенных к телу раскручивает его. Запишем уравнения движения твердого тела:

$$
m \vec{\delta v_{c}} = \vec{F} {\delta \tau} \\
I \frac{\delta \omega}{\delta \tau} = -F \cos \phi R
$$

Посчитаем новые скорости:

$$
v_{cx} = v_{cx0} \\
v_{cy} = v_{cy0} + \frac{\delta p}{m} \\
w = w_0 - \frac{\delta p \cos \phi}{I} R
$$

Запишем закон сохранения энергии

$$
  E = \frac{m (v_{cx}^2 + v_{cy}^2)}{2} + \frac{I \omega^2}{2} = \frac{m (v_{cx0}^2 + v_{cy0}^2)}{2} + \frac{I \omega_{0}^2}{2}
$$

После подстановки можно выразить $\delta p$:

$$
    \delta p = \frac{\omega_0 R \cos \phi - v_{cy0}}{\frac{1}{2 m} + \frac{R^2 \cos^2\phi}{2 I}}
$$

Если присмореться, то окажется что эффект не зависит от массы тела, как и ожидается при упругом столкновении.
